### Example of using bispectrum module

```python
import sys, numpy as np
import tqdm, time # to check progress and elapsed time
from matplotlib.pyplot import *
sys.path.append('../wrap/') # to look for bispec.py 
import bispec  # bispectrum module
```

### Loading matter power spectrum

The matter power spectrum is pre-computed with the same cosmology used in the simulation

```python
D = '../data/' # data directory of pre-computed matter power spectrum at z=0 (from CAMB)
k, pk0 = np.loadtxt(D+'/Pklin.dat',unpack=True) # load the Table of the matter power spectrum (k,Pk) 
```

```python
# plot matter power spectrum
xlabel('$k$[h/Mpc]')
ylabel('$P(k)$[Mpc/h]$^3$')
xscale('log')
yscale('log')
plot(k,pk0)
```

### Computing lensing bispectrum

The lensing bispectrum function below returns $B_{\ell_1\ell_2\ell_3}/h_{\ell_1\ell_2\ell_3}$ where $h_{\ell_1\ell_2\ell_3}$ is defined in Eq.(1.4) of the note

```python
# fitting formula of the matter bispectrum
model = 'RT'
#model = 'GM'
```

```python
#lmax = 2048 # max(l1,l2,l3)
lmax = 1000
#zs = [1.,1.,1.] # source redshifts of three kappa
zs = [1100.,1100.,1100.]
```

Here the number of points for z-integral (zn) is changed for computing the equilateral lensing bispectrum (blll) to check how many z points are enough to compute the lensing bispectrum

```python
# check equilateral
zns  = np.array([10,30,50,100])
zcut = np.array([10,20,40])
blss = np.zeros((len(zns),len(zcut),lmax+1))
bpbc = np.zeros((len(zns),len(zcut),lmax+1))
# loop over zn
for zi, zn in enumerate(zns):
    for ci, zc in enumerate(zcut):
        b1d, b2d, b3d, pb_wp, pb_ck = bispec.bispec_lens_init(lmax,zs,k,pk0,zn=zn,zmaxcut=zc)
        print(np.shape(b1d),np.shape(b2d),np.shape(b3d),np.shape(pb_ck),np.shape(pb_wp))
        # loop over l1
        for l1 in tqdm.tqdm(range(2,lmax)):
            blss[zi,ci,l1] = bispec.bispec_lens(l1,l1,l1,b1d,b2d,b3d,pb_wp,pb_ck,pb=False,zn=zn)
            bpbc[zi,ci,l1] = bispec.bispec_lens(l1,l1,l1,b1d,b2d,b3d,pb_wp,pb_ck,zn=zn)
```

```python
# plot lensing bispectrum for different values of zn 
l = np.linspace(0,lmax,lmax+1)
for zi, zn in enumerate(zns):
    xscale('log')
    yscale('log')
    xlabel('$\ell$')
    ylabel('$\ell^4B_{\ell\ell\ell}/(2\pi)^2/h_{\l\l\l}$')
    xlim(1,lmax)
    for ci, zc in enumerate(zcut):
        bls = blss[zi,ci,:]
        bpb = bpbc[zi,ci,:]
        L = l[bls!=0.]
        Blll = bls[bls!=0.]
        Bpbc = bpb[bls!=0.]
        plot(L,L**4*Blll/(2*np.pi),label='zn='+str(zn)+', zmaxcut='+str(zc))
        plot(L,L**4*Bpbc/(2*np.pi),label='zn='+str(zn)+', zmaxcut='+str(zc)+', post-Born')
    legend()
    show()
    clf()
```

```python
blll = np.zeros((4,lmax+1,lmax+1,lmax+1))
#for zi, zn in enumerate([5,10,20]):
for zi, zn in enumerate([5]):
    # precompute arrays needed for bispectrum calculation in ell-loops
    # this precomputation is neccesarry for fast computation
    b1d, b2d, b3d, pb_wp, pb_ck = bispec.bispec_lens_init(lmax,zs,k,pk0,zn=zn,zmaxcut=20)
    # loop over l1,l2,l3
    start = time.time()
    for l1 in tqdm.tqdm(range(2,lmax),desc=str(zn)):
        for l2 in range(l1,lmax):
            for l3 in range(l2,min(l1+l2,lmax)):
                blll[zi,l1,l2,l3] = bispec.bispec_lens(l1,l2,l3,b1d,b2d,b3d,pb_wp,pb_ck,zn=zn)
    end = time.time()
    print('zn'+str(zn)+':',end-start)
```
